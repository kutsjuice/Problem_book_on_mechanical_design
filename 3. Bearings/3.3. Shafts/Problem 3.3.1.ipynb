```python
import matplotlib.pyplot as plt

def plot_with_stem(ax, x, y, step = None):
    if step==None:
        step = int((x.shape[0]/30))
                   
    x_reduced = x[0::step]
    y_reduced = y[0::step]               
    plt.plot(x,y)
    plt.stem(x_reduced,y_reduced, use_line_collection = True, markerfmt = ' ', linefmt = '-')
                   

xx = np.arange(0,10,0.05)
yy = np.sin(xx)

plot_with_stem(plt.gca(),xx,yy)
plt.show()



    
    
```
