Быстроходный вал косозуой зубчатой передачи расположен на шариковых радиальных подшипниках серии 210. 
Тангенциальная сила в зацеплении $F_t^{зп} = 600$ Н. 
Делительный диаметр шестерни $d_1 = 60$ мм. 
Расстояние между опорами $L = 180$ мм. 
Вал приводится в движение шкивом ременой передачи диаметром $D_ш = 120$ мм. 
Натяжение ведомой ветви ремня составляет $F_{t2}^{рп} = 0.15F_{t1}^{рп}$, где  $F_{t1}^{рп}$ --- натяжение ведущей ветви ремня.
Определить значение коэффициента запаса приняв направление действия консольной силы от ременной передачи наихудшим. 
Значение окржных сил определить из момента на валу.
