Барабан ленточного транспортера установлен на приводном валу, закрепленном на сферическоих шариковых подшипниках.
Во время работы, в ленте транспортера возникают силы натяжения $F_t1 = 1500$ Н и $F_t2 = 500$ Н. Барабан приводится в движение от редуктора при помощи муфты с торообразной оболочкой, радиальная жесткость которой может быть оценена по формуле $C_r=8\sqrt[3]{T}$, где $T$ - номинальный момент вращения. Дать оценку возможным реакциям в подшипниках, если не известо в каком направлении действует консольная сила в муфте, однако известно, чтомаксимальное допустимое смещение концов вала редуктора и приводного вала составляет $\delta = 0,15$ мм.
Диаметр барабана $D = 400$ мм
Расстяние между опорами $L = 450$ мм
Длина концевого участка приводного вала $L = 90$ мм
