# Задача 1
Шпилька с номинальным диаметром 12 мм ввёрнута в абсолютно жесткое основание. На неё с зазором надета втулка и притянута гайкой. Внутренний диаметр втулки – 12,5 мм, наружный – 23 мм Усилие затяжки в шпильке равно 8000 Н. Втулка и шпилька изготовлены из стали одной марки. Определите максимальное значение силы F, при котором втулка ещё будет оказывать давление на основание.


## Решение
Надо рассчитать силу F, при которой втулка не будет оказывать нагрузку на основание. Втулка оторвется в тот момент, когда доля F, действующая на втулку, будет больше силы затяжки. Запишем это в виде уравнения:

$$
F_д = F_з,
$$

где $F_д$ - доля F, действующая на втулку (обозначаем индексом д - деталь). Эту силу можно рассчитать рассчитать по формуле.

$$
F_д = (1 - \chi)\cdot F,
$$

где $\chi$ - коэффициент основной нагрузки. Коэффициент основной нагрузки рассчитывается по формуле:

$$
    \chi = \frac{\lambda_в}{\lambda_в + \lambda_д},
$$

где $\lambda_в$ - податливость винта (винтом является шпилька), мм/Н; $\lambda_д$ - податливость детали (втулки), мм/Н.

Податливости рассчитываются по формулам:

$$
\lambda_в = \frac{l}{E_вA_в} = \frac{l}{E_в \cfrac{\pi d_2 ^ 2}{4}}
$$

$$
\lambda_д = \frac{l}{E_дA_д} = \frac{l}{E_д \cfrac{\pi (d_1 ^ 2 - d_0 ^ 2)}{4}}
$$

Здесь $E_в, E_д$ - модули упругости винта и детали, МПа; $l$ - длина винта/втулки, мм (отличаются незначительно, поэтому приравниваем); $d_2$ - средний диаметр винта, мм; $d_1, d_0$ - внешний и внутренний диаметры втулки, мм; $A_в, A_д$ - площади сечений винта и втулки, $мм^2$.

Рассчитываем $\chi$:

$$
\chi = \frac{\lambda_в}{\lambda_в + \lambda_д} = \frac{\cfrac{l}{E_в \cfrac{\pi d_2 ^ 2}{4}}}{\cfrac{l}{E_в \cfrac{\pi d_2 ^ 2}{4}}+\cfrac{l}{E_д \cfrac{\pi (d_1 ^ 2 - d_0 ^ 2)}{4}}} = \frac{d_1 ^ 2 - d_0 ^ 2}{d_1 ^ 2 - d_0 ^ 2 + d_2 ^2}
$$

Выражаем из этих уравнений F:

$$
F = \frac{F_д}{1 - \chi} = \frac{F_з}{\cfrac{d_1 ^ 2 - d_0 ^ 2}{d_1 ^ 2 - d_0 ^ 2 + d_2 ^2}}
$$

```python
import math

# исходные данные
d = 12         # номинальный диаметр резьбы, мм
d_0 = 12.5     # внутренний диаметр втулки, мм
d_1 = 23       # внешний диаметр втулки, мм
F_zat = 8000   # сила затяжки, Н
P = 1.75

# решение
d_2 = d - 0.649 * P
F = F_zat /( (d_1**2 - d_0**2) / (d_1**2 - d_0**2 + d_2**2))

print(f"Сила F - {F:.3} Н")
```
