# Задача 2
К стальной потолочной плите болтом притянут блок, посредством которого поднимают груз массой 150 кг (см рис.). Определите усилие затяжки и минимально допустимый диаметр болта по впадинам резьбы из условия нераскрытия стыка между основанием блока и плитой. Коэффициент запаса по нераскрытию стыка принять равным 2. Коэффициент основной нагрузки равен 0,2. Предел текучести стали, из которой изготовлен болт, 300 МПа. Коэффициент запаса по текучести должен быть не менее 1,5.

```python
# исходные данные
m = 50          # масса, кг
g = 9.87         # ускорение свободного падения, м/с^2
k_nr = 2         # коэффициент запаса по нераскрытию стыка
chi = 0.2        # коэффициент основной нагрузки
sigma_t = 300    # предел текучести материала болта, МПа
n_t = 3          # коэффициент запаса по текучести
```

сюда картинку

## Решение

Находим силу затяжки $F_з$ из условия нераскрытия стыка:

$$
F_з = k_{нр}2mg(1 - \chi) ~,
$$

где $k_{нр}$ - коэффициент запаса по нераскрытию стыка; $m$ - масса поднимаемого груза, кг; $g$ - ускорение свободного падения, $м/c^2$. В этой формуле масса груза умножается на 2, так как на блок действует как вес груза, так и сила, которая его поднимает. 

```python
F_zat = m * g * (1 - chi) * k_nr
print(f"Сила затяжки - {F_zat:.3}, Н")
```

На болт действуют сила затяжки и часть внешней нагрузки. Суммарная сила, приходящаяся на болт $F_б$:

$$
F_б = m g \chi + F_з ~.
$$

При расчёте эквивалентного напряжения в болте нужно учесть его скручивание пря затяжке. Поэтому при умножаем силу затяжки на коэффициент 1,3.

$$
\sigma_{экв} = \frac{mg\chi + 1{,}3F_з}{\cfrac{\pi d_3^2}{4}} ~, \tag{1}
$$

где $\sigma_{экв}$ - эквивалентное напряжение в винте, МПа; $d_3$ - внутренний диаметр винта, мм. 

Далее запишем условие прочности болта:

$$
\sigma_{экв} \leqslant [\sigma] = \frac{\sigma_t}{n_t} \tag{2} ~,
$$

где $[\sigma]$ - допустимое напряжение для материала болта, МПа; $\sigma_t$ - предел текучести материала болта, МПа; $n_t$ - коэффициент запаса по текучести.

Из формул (1) и (2) выразим минимально допустимый $d_3$:

$$
d_3 = \sqrt{\frac{
                     4 n_t (mg \chi + 1{,}3 F_з)
                 }
                 {
                     \pi \sigma_{t}
                 }
           }
$$


```python
import math

d_3 = math.sqrt(4 * n_t * (m * g * chi + 1.3 * F_zat) / (math.pi * sigma_t)) 

print(f'Минимальный внутренний диаметр винта - {d_3:.3} мм')
```
